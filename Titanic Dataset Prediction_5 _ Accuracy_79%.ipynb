```python
import pandas as pd
import numpy as np
import random
from sklearn import preprocessing, model_selection, neighbors
from sklearn.svm import LinearSVC, SVC
```

```python
df_train=pd.read_csv('train.csv')
df_test=pd.read_csv('test.csv')
```

```python
dftrain = df_train.copy(deep=True)
dftest = df_test.copy(deep=True)
```

```python
dftrain['Age'].isnull().sum()
```

```python
dftest.isnull().sum()
```

```python
dftrain['Age'].fillna(dftrain['Age'].mean(),inplace=True)
dftest['Age'].fillna(dftest['Age'].mean(),inplace=True)
```

```python
lis=dftrain['Embarked'].value_counts().to_dict()
lis[np.nan]=lis['S']
for i in lis:
    lis[i]=lis[i]/dftrain.shape[0]
lis
dftrain['Embarked']=dftrain['Embarked'].map(lis)


liss=dftest['Embarked'].value_counts().to_dict()
liss[np.nan]=liss['S']
for j in liss:
    liss[j]=liss[j]/dftest.shape[0]
liss
dftest['Embarked']=dftest['Embarked'].map(liss)
```

```python
lis2={'male':1,'female':0}
dftrain['Sex']=dftrain['Sex'].map(lis2)
dftest['Sex']=dftest['Sex'].map(lis2)
```

```python
features = ["Pclass", "Sex", "SibSp", "Parch", "Embarked", "Age"]
```

```python
dftrainsvc=df_train.copy(deep=True)
dftestsvc=df_test.copy(deep=True)
```

```python
dftrainsvc['Age'].fillna(dftrainsvc['Age'].mean(),inplace=True)
dftestsvc['Age'].fillna(dftestsvc['Age'].mean(),inplace=True)
```

```python
def ages(age):
    if age<=15:
        return 'child'
    elif age>15 and age<55:
        return 'adult'
    elif age>5:
        return 'old'
                   
```

```python
dftrainsvc['Age']=dftrainsvc['Age'].map(lambda age:ages(age))
dftestsvc['Age']=dftestsvc['Age'].map(lambda age:ages(age))
```

```python
dftrainsvc
```

```python
ysvc=dftrainsvc['Survived']
Xsvc= pd.get_dummies(dftrainsvc[features])
print(Xsvc)
X_testsvc = pd.get_dummies(dftestsvc[features])
print(X_testsvc)
```

```python
model = SVC()
model.fit(Xsvc, ysvc)
predictionsvc = model.predict(X_testsvc)


output = pd.DataFrame({'PassengerId': dftestsvc.PassengerId, 'Survived': predictionsvc})
output.to_csv('Titanic-svc.csv', index=False)
```
